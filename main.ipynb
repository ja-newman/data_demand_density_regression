### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### CSV to DF and Plots

```python
demand_data = pd.read_csv("data_generation_annual.csv", index_col=0, parse_dates=True)
density_data = pd.read_csv("data_storage_density_annual.csv", index_col=0, parse_dates=True)

plt.plot(demand_data, color='r', label='Demand (Zettabytes)')
plt.plot(density_data, color='g', label='Density (TB/Sqft)')

plt.xlabel("Year")
plt.ylabel("Magnitude")
plt.title("Data collection and storage density trends")

plt.legend()
plt.show()
```
